```python
df1=df.copy()
for i in df.columns:
    if df[i].dtypes!='object':
        q1=df[i].quantile(0.25)
        q3=df[i].quantile(0.75)
        iqr=q3-q1
        lower_fence=q1-1.5*iqr
        uper=q3+1.5*iqr
        
        df1 = df1[(df1[i] >= lower_fence) & (df1[i] <= uper)]
```
